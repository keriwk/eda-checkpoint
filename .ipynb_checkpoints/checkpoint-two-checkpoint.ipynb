```python
#evaluate correlation between variables
# select only numerical columns
numerical_columns = df_activity.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numerical_columns.corr()
print(correlation_matrix)
```

```python
numerical_columns = df_sleep.select_dtypes(include=['float64', 'int64'])

correlation_matrix = numerical_columns.corr()
print(correlation_matrix)
```

```python
# convert date and split date and time to help with any join on dates
df_sleep['SleepDay'] = pd.to_datetime(df_sleep['SleepDay'], format = '%m/%d/%Y %I:%M:%S %p')
df_sleep['Date'] = df_sleep['SleepDay'].dt.date
df_sleep['Time'] = df_sleep['SleepDay'].dt.time
df_sleep.head()
```

```python
# change ActivityDate to Date type and add 'Date' for inner join
df_activity['ActivityDate'] = pd.to_datetime(df_activity['ActivityDate'])
df_activity['Date'] = df_activity['ActivityDate'].dt.date
df_activity.info()
```

```python
# Inner join daily data
daily_df = pd.merge(df_activity, df_sleep, on = ['Id', 'Date'], how = 'inner')
daily_df.info()
```

```python
daily_df.head()
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#sum different levels of activity minutes to see distribution of total among all participants. More sedentary minutes.

activity_types = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
minutes = [
    daily_df['VeryActiveMinutes'].sum(),
    daily_df['FairlyActiveMinutes'].sum(),
    daily_df['LightlyActiveMinutes'].sum(),
    daily_df['SedentaryMinutes'].sum()
]

plt.figure(figsize=(10, 6))
plt.bar(activity_types, minutes, color=['red', 'orange', 'yellow', 'grey'])
plt.title('Total Minutes Spent in Different Activity Types')
plt.xlabel('Activity Type')
plt.ylabel('Total Minutes')
plt.show()

```

```python
#I want to see if total sleep minutes is correlated to total time in bed, which is seems to be. A few outliers

plt.figure(figsize=(10, 6))
plt.scatter(df_sleep['TotalTimeInBed'], df_sleep['TotalMinutesAsleep'], alpha=0.5)
plt.title('Sleep Minutes vs. Total Time in Bed')
plt.xlabel('Total Time in Bed (minutes)')
plt.ylabel('Total Minutes Asleep')
plt.show()
```

```python
#Comparing total sleep time to total time in bed with averages
# average sleep time and average total time in bed
avg_sleep_time = df_sleep['TotalMinutesAsleep'].mean()
avg_total_time_in_bed = df_sleep['TotalTimeInBed'].mean()

# bar chart
categories = ['Average Sleep Time', 'Average Total Time in Bed']
values = [avg_sleep_time, avg_total_time_in_bed]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=['blue', 'green'])
plt.title('Average Sleep Time vs. Average Total Time in Bed')
plt.xlabel('Category')
plt.ylabel('Minutes')
plt.show()
```
